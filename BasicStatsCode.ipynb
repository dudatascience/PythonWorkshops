```python
f = open('HEIGHT.csv')
heights = []

for line in f.readlines():
    row = line.split(',')
    if row[0] == 'AGE':
        continue
    heights.append(float(row[6]))
```

```python
import numpy
from scipy import stats

print(len(heights))
print(numpy.mean(heights))
print(numpy.median(heights))
print(stats.mode(heights))
print(stats.mode(heights).mode[0])
```

```python
def distance_from_mean(point, dataset):
    mean = numpy.mean(dataset)
    return point - mean
```

```python
distance_from_mean(5, heights)
```

```python
def squared_distance_from_mean(point, dataset):
    return distance_from_mean(point, dataset)**2
```

```python
squared_distance_from_mean(1.65, heights)
```

```python
squared_distance_from_mean(1.75, heights)
```

```python
def average_squared_distance_from_mean(dataset):
    distances_from_mean = []
    for point in dataset:
        distances_from_mean.append(squared_distance_from_mean(point, dataset))
    
    return numpy.mean(distances_from_mean)
```

```python
average_squared_distance_from_mean(heights)
```

```python
numpy.var(heights)
```

```python
import math

def standard_deviation(dataset):
    return math.sqrt(average_squared_distance_from_mean(dataset))
```

```python
standard_deviation(heights)
```
